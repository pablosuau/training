Two of the examples involve a single flip, with the only difference between the examples being whether the prior is uniform or contain only two extreme options. The point of these two examples is to show that a single datum implies little when the prior is vague, buy a single datum can have strong implications when the prior allows only two very different possibilities. 

```python
# Sparse teeth for theta
theta = np.linspace(0 , 1 , 5)  
# Triangular shape for p_theta
p_theta = np.minimum(theta, 1 - theta) 
# Make p_theta sum to 1.0
p_theta = p_theta / np.sum(p_theta)    
# Single flip with 1 head
data = np.concatenate((np.repeat(0, 0), np.repeat(1, 1)), axis = None)

posterior = bern_grid(theta, 
                      p_theta, 
                      data, 
                      plot_type = 'bars', 
                      show_cent_tend = 'none', 
                      show_hdi = False,
                      show_pd = False)
```

```python
# Sparse teeth for theta
theta = np.linspace(0 , 1 , 1000)  
# Triangular shape for p_theta
p_theta = np.zeros(len(theta.tolist()))
p_theta[1] = 1
p_theta[-2] = 1
# Make p_theta sum to 1.0
p_theta = p_theta / np.sum(p_theta)    
# Single flip with 1 head
data = np.concatenate((np.repeat(0, 0), np.repeat(1, 1)), axis = None)

posterior = bern_grid(theta, 
                      p_theta, 
                      data, 
                      plot_type = 'bars', 
                      show_cent_tend = 'none', 
                      show_hdi = False,
                      show_pd = False)
```

However, when the prior only accepts two values, a single datum has a big impact on our believes. 

```python
# Sparse teeth for theta
theta = np.linspace(0 , 1 , 1000)  
# Triangular shape for p_theta
p_theta = np.minimum(theta, 1 - theta) 
# Make p_theta sum to 1.0
p_theta = p_theta / np.sum(p_theta)    
# 25% heads, N = 4
data = np.concatenate((np.repeat(0, 3), np.repeat(1, 1)), axis = None)

posterior = bern_grid(theta, 
                      p_theta, 
                      data, 
                      plot_type = 'bars', 
                      show_cent_tend = 'mode', 
                      show_hdi = True,
                      show_pd = False)
```

The prior has a larger impact on the posterior when the sample size is small (see next).

```python
# Sparse teeth for theta
theta = np.linspace(0 , 1 , 1000)  
# Triangular shape for p_theta
p_theta = np.minimum(theta, 1 - theta) 
# Make p_theta sum to 1.0
p_theta = p_theta / np.sum(p_theta)
# 25% heads, N = 40
data = np.concatenate((np.repeat(0, 30), np.repeat(1, 10)), axis = None)

posterior = bern_grid(theta, 
                      p_theta, 
                      data, 
                      plot_type = 'bars', 
                      show_cent_tend = 'mode', 
                      show_hdi = True,
                      show_pd = False)
```

The prior has a smaller impact on the posterior when the sample size is large (see previous).

```python
# Sparse teeth for theta
theta = np.linspace(0 , 1 , 1000)  
# Triangular shape for p_theta
p_theta = np.minimum(theta, 1 - theta) 
# Make p_theta sum to 1.0
p_theta = p_theta / np.sum(p_theta) 
# Sharpen p_theta
p_theta = np.power(p_theta, 10)
# Make p_theta sum to 1.0
p_theta = p_theta / np.sum(p_theta)
# 25% heads, N = 4
data = np.concatenate((np.repeat(0, 3), np.repeat(1, 1)), axis = None)

posterior = bern_grid(theta, 
                      p_theta, 
                      data, 
                      plot_type = 'bars', 
                      show_cent_tend = 'mode', 
                      show_hdi = True,
                      show_pd = False)
```

The prior has a large impace on the posterior if it is sharp (see next).

```python
# Sparse teeth for theta
theta = np.linspace(0 , 1 , 1000)  
# Triangular shape for p_theta
p_theta = np.minimum(theta, 1 - theta) 
# Make p_theta sum to 1.0
p_theta = p_theta / np.sum(p_theta) 
# Flatten p_theta
p_theta = np.power(p_theta, 0.1)
# Make p_theta sum to 1.0
p_theta = p_theta / np.sum(p_theta)
# 25% heads, N = 4
data = np.concatenate((np.repeat(0, 3), np.repeat(1, 1)), axis = None)

posterior = bern_grid(theta, 
                      p_theta, 
                      data, 
                      plot_type = 'bars', 
                      show_cent_tend = 'mode', 
                      show_hdi = True,
                      show_pd = False)
```

As the size of the sample increases, the posterior is calculated as a trade-off between the prior and the likelihood. In this case, even though the prior is very sharp, the likelihood has an impact on the posterior. 

```python
# Sparse teeth for theta
theta = np.linspace(0 , 1 , 1000)  
# Triangular shape for p_theta
p_theta = np.minimum(theta, 1 - theta) 
# Make p_theta sum to 1.0
p_theta = p_theta / np.sum(p_theta) 
# Flatten p_theta
p_theta = np.power(p_theta, 0.1)
# Make p_theta sum to 1.0
p_theta = p_theta / np.sum(p_theta)
# 25% heads, N = 40
data = np.concatenate((np.repeat(0, 30), np.repeat(1, 10)), axis = None)

posterior = bern_grid(theta, 
                      p_theta, 
                      data, 
                      plot_type = 'bars', 
                      show_cent_tend = 'mode', 
                      show_hdi = True,
                      show_pd = False)
```
