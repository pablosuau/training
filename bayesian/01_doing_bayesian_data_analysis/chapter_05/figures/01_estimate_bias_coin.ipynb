```python
import os, sys
import numpy as np
util_dir = os.path.dirname('../../util/')
sys.path.insert(0, util_dir) 
from bern_grid import bern_grid
```

**Figure 5.1** Bayes' rule applied to estimating the bias of a coin. There are discrete value candidates of $\theta$. At each value of $\theta$, the posterior is computed as prior times likelihood, normalised. In the data, denoted $D$, the number of heads is $z$ and the number of flips is $N$. in this case $z = 1$, $N = 1$.

```python
# Sparse teeth for theta
theta = np.linspace(0 , 1 , 11)  
# Triangular shape for p_theta
p_theta = np.minimum(theta, 1 - theta) 
# Make p_theta sum to 1.0
p_theta = p_theta / np.sum(p_theta)    
# Single flip with 1 head
data = np.concatenate((np.repeat(0, 0), np.repeat(1, 1)), axis = None)

posterior = bern_grid(theta, 
                      p_theta, 
                      data, 
                      plot_type = 'bars', 
                      show_cent_tend = 'none', 
                      show_hdi = False,
                      show_pd = False)
```

**Figure 5.2** The two columns show different sample sizes with the same proportion of heads. The prior is the same in both columns. The influence of the prior is overwhelmed by larger samples, in that the peak of the posterior is closer to the peak of the likelihood function. Notice also that the posterior HDI is narrower for the larger sample. 

```python
# Sparse teeth for theta
theta = np.linspace(0 , 1 , 1000)  
# Triangular shape for p_theta
p_theta = np.minimum(theta, 1 - theta) 
# Make p_theta sum to 1.0
p_theta = p_theta / np.sum(p_theta)    
# 25% heads, N = 4
data = np.concatenate((np.repeat(0, 3), np.repeat(1, 1)), axis = None)

posterior = bern_grid(theta, 
                      p_theta, 
                      data, 
                      plot_type = 'bars', 
                      show_cent_tend = 'mode', 
                      show_hdi = True,
                      show_pd = False)
```

**Figure 5.3** The left side is teh same small sample as the left side of Figure 5.2 but with a flatter prior. The right side is the same larger sample as the right side of Figure 5.2 but with a sharper prior. 

```python
# Sparse teeth for theta
theta = np.linspace(0 , 1 , 1000)  
# Triangular shape for p_theta
p_theta = np.minimum(theta, 1 - theta) 
# Make p_theta sum to 1.0
p_theta = p_theta / np.sum(p_theta) 
# Flatten p_theta
p_theta = np.power(p_theta, 0.1)
# Make p_theta sum to 1.0
p_theta = p_theta / np.sum(p_theta)
# 25% heads, N = 4
data = np.concatenate((np.repeat(0, 3), np.repeat(1, 1)), axis = None)

posterior = bern_grid(theta, 
                      p_theta, 
                      data, 
                      plot_type = 'bars', 
                      show_cent_tend = 'mode', 
                      show_hdi = True,
                      show_pd = False)
```
