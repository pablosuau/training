**Exercise 6.1. [Purpose: For you to see the influence of the prior in each successive flip, and for you to see another demonstration that the posterior is invariant under re-orderings of the data.]** For this exercise, use bern_beta. Notice that the function returns the posterior beta values each time it is called, so you can use the returned values as the prior values for the next function call. 

**(A)** Start with a prior distribution that expresses some uncertainty that a coin is fair: $beta(\theta|4, 4)$. Flip the coin once; suppose that we get a head. What is the posterior distribution?

```python
a = 4
b = 4

prior = [a,b] # Specify prior as vector with the two shape parameters.

# Specify the data:
N = 1 # The total number of flips.
z = 1 # The number of heads.
# Convert N and z into vector of 0's and 1's.
head = np.array([1])

posterior = bern_beta(prior_beta_ab = prior, 
                      data = head , 
                      plot_type = 'bars', 
                      show_cent_tend = 'mode' , 
                      show_hdi = True, 
                      show_pd = False)
```

```python
posterior
```

The posterior distribution is $beta(\theta|5, 4)$

**(B)** Use the posterior from the previous flip as the prior for the next flip. Suppose that we flip again and get a head. Now, what is the new posterior?

```python
posterior = bern_beta(prior_beta_ab = posterior, 
                      data = head , 
                      plot_type = 'bars', 
                      show_cent_tend = 'mode' , 
                      show_hdi = True, 
                      show_pd = False)
```

```python
posterior
```

The posterior distribution is  $beta(\theta|6,4)$

**(C)** Using that posterior as the prior for the next flip, flip a third time and get a tail. Now what is the new posterior?

```python
tail = np.array([0])

posterior = bern_beta(prior_beta_ab = posterior, 
                      data = tail , 
                      plot_type = 'bars', 
                      show_cent_tend = 'mode' , 
                      show_hdi = True, 
                      show_pd = False)
```

```python
posterior
```

The new posterior is $beta(\theta|6, 5)$

**(D)** Do the same three updates but in the order T, H, H instead of H, H, T. Is the final posterior distribution the same for both orderings of the flip results?

```python
posterior = bern_beta(prior_beta_ab = prior, 
                      data = tail , 
                      plot_type = 'bars', 
                      show_cent_tend = 'mode' , 
                      show_hdi = True, 
                      show_pd = False)

posterior = bern_beta(prior_beta_ab = posterior, 
                      data = head , 
                      plot_type = 'bars', 
                      show_cent_tend = 'mode' , 
                      show_hdi = True, 
                      show_pd = False)

posterior = bern_beta(prior_beta_ab = posterior, 
                      data = head , 
                      plot_type = 'bars', 
                      show_cent_tend = 'mode' , 
                      show_hdi = True, 
                      show_pd = False)
```

```python
posterior
```

Yes, the posterior is exactly the same no matter the order of the observations.
