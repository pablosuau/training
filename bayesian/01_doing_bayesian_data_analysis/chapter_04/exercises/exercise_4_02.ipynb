```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(47408)
```

```python
N = 500
p_heads = 0.8
# Generate a random sample of N flips (heads = 1, tails = 0)
flip_sequence = np.random.binomial(1, p_heads, size = N)
# Compute the running proportion of heads
r = np.cumsum(flip_sequence)
n = np.arange(1, N + 1)
run_prop = r / n
# Graph the running proportion
plt.plot(n, run_prop, 'o-', alpha = 0.5)
plt.xscale('log')
plt.xlim([1, N])
plt.ylim([0, 1])
plt.xlabel('flip number')
plt.ylabel('proportion heads')
plt.title('running proportion of heads')
# Plot a dotted horizontal reference line
plt.axhline(y = 0.8, linestyle = ':', color = 'r', alpha = 0.5)
# Display the beginning of the plot sequence
flip_letters = np.array2string(flip_sequence[:10], separator = '')[1:11].replace('1', 'H').replace('0', 'T')
plt.text(6, 0.25, 'Flip sequence = ' + flip_letters + '...', size = 14)
# Display the relative frequency at the end of the sequence
_ = plt.text(20, 0.15, 'End proportion = ' + str(run_prop[-1]), size = 14)
```
