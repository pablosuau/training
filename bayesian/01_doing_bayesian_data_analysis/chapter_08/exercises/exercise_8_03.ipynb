My Python version of the script is not saving any data. Therefore, I looked into the original R code to answer this question:

* `genMcmc` saves the trace, that is, the sampled values for the `theta_1` and `theta_2` parameters.
* `smryMCMC` saves the summary information displayed in the exercise 8 02 notebook as a CSV file
* `plotMCMC` saves the figure generated by this function: posterior for the `theta_1`, `theta_2` and `theta_1 - theta_2` parameters as well as the scatter plot of sampled `theta_1` vs `theta_2` values at each iteration.
